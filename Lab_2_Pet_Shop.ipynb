## Lab 2: Pet Shop Entity Relation Diagram

Lab 2 Pet Shop ER Diagram.png

#### This diagram attempts to create the outline for a basic database intended for a small pet shop. Many pets are kept in many enclosures. The pet store may have up to 1000 pets at a time with a variety of large and small pets alike. The pet store has up to 100 different enclosures suited for specific pets based on the species of the pet. Food is fed to the pets based upon the diet of any particular pet. So, many types of food are fed to many pets. The pet store possesses up to 50 varieties of food for the pets. Customers are able to purchase both pets and food. Many customers purchase many pets. The store may have up to 10,000 customers. Many customers may also purchase many food products. Pets are sold through a transaction based upon the stock code of the pet. Up to ten pets may be sold per single transaction. The same is true for food. Food is sold through a transaction based upon the stock code of that food item. Up to ten food items may be sold per single transaction. A customer loyalty program also exists. Customers have a single loyalty ID. Customers also have Awards Points. Awards points are issured to the customers through transactions and the Loyalty program. Awards points are used to determine Loyalty Awards. Loyalty Awards are applied to discount transactions.
